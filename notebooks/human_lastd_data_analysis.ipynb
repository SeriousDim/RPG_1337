```python
from pathlib import Path

import pandas as pd

import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("./results_25-04-2026_dataframe.csv")
```

```python
len(df)
```

```python
df['session_id'].unique()
```

```python
df["session_id"] = df["session_id"].fillna("99998888")
```

```python
counts = df["session_id"].value_counts(dropna=False)

plt.figure(figsize=(10, 5))
counts.plot(kind="bar")
plt.xlabel("session_id")
plt.ylabel("Количество записей")
plt.title("Количество записей по session_id")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
